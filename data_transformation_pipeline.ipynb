```python
import os
import numpy as np
import pandas as pd
```

# Function to transform/view data

```python
#################################################
# Function to transform/view data
#################################################
def files_to_df(data_folder: str, view_status: bool, no_to_process: int, convert_to_csv: bool):
    data_files = os.listdir(data_folder)
    
    for file in data_files[:no_to_process]:
        dfs = os.listdir(os.path.join(data_folder, file))
        if view_status:
            print(file, ":")
        
        for df in dfs:
            df_path = os.path.join(data_folder, file)
            df_path = os.path.join(df_path, df)
            
            if not os.path.exists(df_path):
                print("File does not exist : ", df_path)
                continue
            
            if view_status:
                print("      ->",df_path)
            
            dataframe = pd.read_xml(df_path)
```

# Call the function to transform/view data
